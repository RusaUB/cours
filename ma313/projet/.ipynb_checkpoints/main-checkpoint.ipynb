## 1 Approximation de fonctions par des polynômes

### 1 
N représente le nombre de points de données disponibles à partir de la fonction. Pour ajuster un polynôme de degré d, il est nécessaire d'avoir suffisamment de points (N) pour que le polynôme puisse être correctement ajusté. 
Pour un ajustement précis, il est nécessaire d'avoir un nombre de points N plus grand que le nombre de degrés de liberté d du polynôme. Sinon, le système peut être surcontraint ou instable. Plus il y a de points de données, plus le polynôme peut capturer les variations de la fonction.

### 2
Pour un polynôme de degré d, le problème aux moindres carrés consiste à trouver les coefficients $c_0,c_1,...,c_d$ du polynôme $P(x) = c_0 + c_1x + c_2x + ... + c_dx^d$ tels que la somme des carrés des différences entre $P(x_i)$ et les valeurs observées $y_i$ soit minimale pour N points de données $(x_1,y_1),(x_2,y_2),...,(x_N,y_N)$

$$
\begin{bmatrix}
    1 & x_1 & x_1^2 & \cdots & x_1^d \\
    1 & x_2 & x_2^2 & \cdots & x_2^d \\
    \vdots & \vdots & \vdots & \ddots & \vdots \\
    1 & x_N & x_N^2 & \cdots & x_N^d \\
\end{bmatrix}
\begin{bmatrix}
    c_0 \\
    c_1 \\
    \vdots \\
    c_d \\
\end{bmatrix}
=
\begin{bmatrix}
    y_1 \\
    y_2 \\
    \vdots \\
    y_N \\
\end{bmatrix}
$$

Le noyau d'une matrice A est défini comme : 

$
\text{Ker}(A) = \{ \mathbf{x} \in \mathbb{R}^n : A\mathbf{x} = \mathbf{0} \}
$

### 4

$$
\begin{bmatrix}
    1 & x_1 & x_1^2 & \cdots & x_1^d \\
    1 & x_2 & x_2^2 & \cdots & x_2^d \\
    \vdots & \vdots & \vdots & \ddots & \vdots \\
    1 & x_N & x_N^2 & \cdots & x_N^d \\
\end{bmatrix}
\begin{bmatrix}
    c_0 \\
    c_1 \\
    \vdots \\
    c_d \\
\end{bmatrix}
=
\begin{bmatrix}
    e^{x_1} \\
    e^{x_2} \\
    \vdots \\
    e^{x_N} \\
\end{bmatrix}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

def f1(x):
    return np.exp(x)

def solve_normal_equations(x, f, d):
    A = np.vander(x, d + 1, increasing=True)
    return np.linalg.lstsq(A, f, rcond=None)[0]

def plot_approximation_function(func, degrees=list(range(12)), show_heat_map=False):
    # Generate x values
    N_values = [50, 100]  # Different values of N (N = d + 1)
    x_values = np.linspace(-1, 1, max(N_values))

    # Calculate f(x) values using the provided function
    f_values = func(x_values)

    num_cols = 3  # Maximum number of columns per row

    num_plots = len(degrees)
    num_rows = -(-num_plots // num_cols)  # Calculate the number of required rows

    fig, axs = plt.subplots(num_rows, num_cols, figsize=(15, 5 * num_rows))
    axs = axs.flatten()

    max_errors = []  
    condition_numbers = [] 

    for i, d in enumerate(degrees):
        ax = axs[i]
        for N in N_values:
            coefficients_normal = solve_normal_equations(x_values[:N], f_values[:N], d)
            poly_normal = np.poly1d(coefficients_normal[::-1])

            error = np.max(np.abs(f_values[:N] - poly_normal(x_values[:N])))
            max_errors.append(error)

            A = np.vander(x_values[:N], d + 1, increasing=True)
            condition_number = np.linalg.cond(A)
            condition_numbers.append(condition_number)

            ax.plot(x_values, poly_normal(x_values), label=f'N={N} (Équations normales)')

        ax.plot(x_values, f_values, label='f(x)', linestyle='--', color='black')
        ax.grid()
        ax.set_xlabel('x')
        ax.set_ylabel('f(x)')
        ax.set_title(f'Approximation for d = {d}')
        ax.legend()

    max_errors = np.array(max_errors).reshape(len(degrees), len(N_values))
    condition_numbers = np.array(condition_numbers).reshape(len(degrees), len(N_values))

    fig, axes = plt.subplots(1, 2, figsize=(15, 6))
    titles = ['Max Error |f(x) - P(x)|', 'Condition Number of the System']

    for i, data in enumerate([max_errors, condition_numbers]):
        ax = axes[i]
        im = ax.imshow(data, cmap='viridis', aspect='auto')
        ax.set_xlabel('N Values')
        ax.set_ylabel('Polynomial Degree (d)')
        ax.set_xticks(np.arange(len(N_values)))
        ax.set_xticklabels(N_values)
        ax.set_yticks(np.arange(len(degrees)))
        ax.set_yticklabels(degrees)
        ax.set_title(titles[i])
        plt.colorbar(im, ax=ax)

    plt.tight_layout()
    plt.show()
```

```python
plot_approximation_function(f1)
```

### 5

```python
plot_approximation_function(func=f1,degrees=[15,20,25,30,35,40])
```

### 6

```python
def f2(x):
    term_den = (1+x**2)
    return 1/term_den

plot_approximation_function(func=f2)
```
