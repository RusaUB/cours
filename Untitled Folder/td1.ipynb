```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,100)
f1 = np.abs(x)

plt.plot(x,f1)
plt.show()
```

```python
f2 =np.exp(x) - x

plt.plot(x,f2)
plt.show()
```

```python
f3 =2*x**2 + x + 1


plt.plot(x,f3)
plt.show()
```

```python
f3 =4*x + 1


plt.plot(x,f3)
plt.show()
```

```python
f5 = 2*((x-1)**4) + 1


plt.plot(x,f5)
plt.show()
```

```python
f6 = x**3 + x + 1


plt.plot(x,f6)
plt.show()
```

```python
f7 = []

for i in x:
    if i >= 0:
        f7.append(i**2)
    else :
        f7.append(0)

plt.plot(x,f7)
plt.show()
```

```python
f8 = []

for i in x:
    if i >= 0:
        f8.append(i**3)
    else :
        f8.append(0)

plt.plot(x,f8)
plt.show()
```
